```python
import matplotlib.pyplot as plt
import numpy as np

Xpoint=np.array([0,6])
Ypoint=np.array([0,250])

plt.plot(Xpoint, Ypoint) 
plt.show()
```

```python
plt.plot(Xpoint,Ypoint,'o')
```

```python
Xpoint=np.array([1,5,8,5])
Ypoint=np.array([2,4,6,3])
plt.plot(Xpoint,Ypoint,	marker='o',linestyle="--")
plt.show()
```

```python
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("x-y plot")
plt.plot(Xpoint,Ypoint,	marker='o',linestyle="--")
plt.show()
```

```python
font1={'family':'serif','color':'blue','size':10}
font2={'family':'monospace','color':'blue','size':10}
font3={'family':'monospace','color':'red','size':20}

plt.xlabel("x-values",fontdict=font1)
plt.ylabel("y-values",fontdict=font2)
plt.title("x-y plot" ,fontdict=font3)
plt.plot(Xpoint,Ypoint,	marker='o',linestyle="--")
plt.show()
```

```python
plt.plot(Xpoint,Ypoint)
# plt.grid()
# plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```

```python
plt.plot(Xpoint,Ypoint,marker='o',linestyle="--")
x2=np.array([1,9,4,6])
y2=np.array([3,5,7,8])
plt.plot(x2,y2,marker='o',linestyle="--")
plt.show()
```
